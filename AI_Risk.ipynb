```python
import pandas as pd

# Sirf un rows ko rakho jinka AI risk 70% ya usse zyada hai
high_risk = df[df["AI_Replacement_Risk"] >= 0.90]

# Count store karne ke liye
count = {}

# High-risk rows par loop
for row in high_risk["Satisfaction|Hours|Remote_work|Company_siz|Salary_range"]:

    # Combination ko split karo
    values = row.split("|")

    # Har value ka count badhao
    for value in values:

        if value not in count:
            count[value] = 1
        else:
            count[value] += 1

# Table banake descending order mein sort karo
result = pd.DataFrame(
    count.items(),
    columns=["Value", "Count"]
).sort_values("Count", ascending=False)

print(result)
```

```python
import pandas as pd

# Start from your 76-row dataframe
combination_df = final_df.copy()   # or use the dataframe that contains the 76 rows

# --------------------------------------------------
# Step 1: OR filter
# --------------------------------------------------
filtered = combination_df[
    (combination_df['Education_Level'] == 'Master') |
    (combination_df['Exp_range'] == 'Lead') |
    (combination_df['job_gro_range'] == 'Rapid Growth') |
    (combination_df['Hiring_Trend_2026'] == 'Growing') |
    (combination_df['Hours_range'] == 'Heavy overtime') |
    (combination_df['Upskilling_Needed'] == 'No')
].copy()

# --------------------------------------------------
# Step 2: Average AI risk by Combination 1 column
# --------------------------------------------------
combo_col = 'Satisfaction|Hours|Remote_work|Company_siz|Salary_range'

avg_risk = (
    filtered
    .groupby(combo_col)['AI_Replacement_Risk']
    .mean()
    .reset_index(name='avg_ai_risk')
)

# --------------------------------------------------
# Step 3: Keep combinations with avg AI risk >= 0.7
# --------------------------------------------------
valid_groups = avg_risk.loc[
    avg_risk['avg_ai_risk'] >= 0.7,
    combo_col
]

# --------------------------------------------------
# Step 4: Filter rows belonging to valid combinations
# --------------------------------------------------
final_df = filtered[
    filtered[combo_col].isin(valid_groups)
].copy()

# --------------------------------------------------
# Step 5: Frequency analysis
# --------------------------------------------------
target_cols = [
    'Remote_Work_Possibility',
    'Automation_Level',
    'AI_Tool_Usage',
    'perfomance',
    'Satis_range',
    'Demand_range',
    'Industry',
    'Salary_range'
]

freq_tables = []

for col in target_cols:
    counts = (
        final_df[col]
        .value_counts(dropna=False)
        .reset_index()
    )
    counts.columns = [col, 'count']
    freq_tables.append(counts)

result = pd.concat(freq_tables, axis=1)

# --------------------------------------------------
# Output
# --------------------------------------------------
print("\nAverage AI Risk by Combination:")
print(avg_risk.sort_values('avg_ai_risk', ascending=False))

print("\nRows after AI Risk >= 0.7 filter:")
print(len(final_df))

print("\nFrequency Summary:")
print(result)
```

```python
import pandas as pd

combo_col = 'Satisfaction|Hours|Remote_work|Company_siz|Salary_range'

# =====================================================
# STEP 1: Average AI risk by combination
# =====================================================

avg_risk = (
    df.groupby(combo_col)['AI_Replacement_Risk']
      .mean()
      .reset_index(name='avg_ai_risk')
)

# =====================================================
# STEP 2: Keep combinations with avg risk >= 0.7
# =====================================================

valid_combos = avg_risk.loc[
    avg_risk['avg_ai_risk'] >= 0.7,
    combo_col
]

risk_filtered = df[
    df[combo_col].isin(valid_combos)
].copy()

# =====================================================
# STEP 3: Apply OR filter
# =====================================================

final_df = risk_filtered[
    (risk_filtered['Education_Level'] == 'Master') |
    (risk_filtered['Exp_range'] == 'Lead') |
    (risk_filtered['job_gro_range'] == 'Rapid Growth') |
    (risk_filtered['Hiring_Trend_2026'] == 'Growing') |
    (risk_filtered['Hours_range'] == 'Heavy overtime') |
    (risk_filtered['Upskilling_Needed'] == 'No')
].copy()

# =====================================================
# STEP 4: Frequency counts
# =====================================================

target_cols = [
    'Job_Title', 
    'Industry', 
    'Country'
]

tables = []

for col in target_cols:

    temp = (
        final_df[col]
        .value_counts(dropna=False)
        .reset_index()
    )

    temp.columns = [col, f'{col}_count']

    tables.append(temp)

# Side-by-side table
result = pd.concat(tables, axis=1)

# =====================================================
# OUTPUT
# =====================================================

print("Rows after risk filter + OR filter:", len(final_df))
print(result)
```
